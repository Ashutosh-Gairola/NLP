# Program to find the word "the" in the brownÂ corpus

```python
import re
import nltk
nltk.download('all')
```

```python
from nltk.corpus import brown
```

```python
brown_corpus=brown.sents()
all_words=[words.lower() for sentence in brown_corpus for words in sentence]
pattern=r'\bthe\b'
```

```python
instances_of_the=[word for word in all_words if re.match(pattern,word)]
```

```python
print("instances:",instances_of_the)
```

```python
count_len=len(instances_of_the)
print("Number of the instances of 'the' is:",count_len)
```

# Program to print all the bigrams and trigrams from a given corpus and sort themÂ byÂ frequency

```python
from nltk import bigrams,trigrams
```

```python
brown_corpus=brown.sents()
all_words=[words.lower() for sentence in brown_corpus for words in sentence]
```

```python
all_bigram=list(bigrams(all_words))
all_trigram=list(trigrams(all_words))
```

```python
print("Bigrams:",all_bigram[:10])
```

```python
print("Trigrams:",all_trigram[:10])
```

```python
import fitz
```

```python
pdf_path="/content/drive/MyDrive/Colab Notebooks/NLP_Sem7_pythonLab/the-enchanted-forest-adventure.pdf"
# pdf_path="/content/drive/MyDrive/Colab Notebooks/NLP_Sem7_pythonLab/SeventhSense-obooko-hor0070.pdf"
# pdf_path="/content/drive/MyDrive/Colab Notebooks/NLP_Sem7_pythonLab/short-stories-and-other-writings-obooko.pdf"
```

```python
def extract_text_from_pdf(pdf_path):
    doc=fitz.open(pdf_path)
    text=""
    for page_num in range(doc.page_count):
        page=doc.load_page(page_num)
        text+=page.get_text()
    doc.close()
    return text

```

```python
pdf_text= extract_text_from_pdf(pdf_path)
```

```python
tokens=word_tokenize(pdf_text)
print(tokens)
```

```python
whitespace_tokens=pdf_text.split()
print(whitespace_tokens)
```

# Sentence Tokenizer

```python
from nltk.tokenize import PunktSentenceTokenizer
```

```python
tokenizer=PunktSentenceTokenizer()
tokenizer.train(pdf_text)

punkt_tokens=tokenizer.tokenize(pdf_text)
```

```python
for index, sentence in enumerate(punkt_tokens, start=1):
    print(f"Sentence {index}: {sentence}\n")
```
